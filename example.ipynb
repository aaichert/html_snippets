```python
from html_snippets.image import html_image
print('This image has been converted to html code for diplay:')
HTML(html_image(image))
```

```python
import PIL.ImageOps
# Create an inverted image to demonstrate overlay feature.
inverted_image = PIL.ImageOps.invert(image)
```

```python
from html_snippets.overlay import html_auto_blend_last_child
print('you can use this to compare two images or hos/hide an overlay')
HTML(' '.join(html_auto_blend_last_child(html_image(image)+html_image(inverted_image))))
```

```python
from html_snippets.overlay import html_overlay_hide_on_hover
print('hover mouse to compare images:')
HTML(' '.join(html_overlay_hide_on_hover(html_image(image)+html_image(inverted_image))))
```

```python
from html_snippets.table import html_table
```

```python
# As dummy data, crete a table of image crops
import uuid
import random
random.seed(1337)

crop_size = 32

table_data = []
for _ in range(5):
    uid = str(uuid.uuid4())[:8]
    x = random.randint(0, image.width - crop_size)
    y = random.randint(0, image.height - crop_size)
    cropped_image = image.crop((x, y, x + crop_size, y + crop_size))
    table_data.append({'uid': uid, 'image': cropped_image, 'x': x, 'y': y})

print("Table data containing PIL.Image objects")
print(table_data)
print()

# Convert images from PIL.Image to a string with HTML code
for row in table_data:
    row['image'] = html_image(row['image'])
```

```python
# Create table HTML
head, body, tail = html_table(table_data, simple=True)
HTML(head+body+tail)
```

```python
from html_snippets.table import sort_rows, color_code, rename_column
```

```python
sort_rows(table_data, 'x')
color_code(table_data, 'x', 0, 92)
color_code(table_data, 'y', 0, 92)
rename_column(table_data, 'x', 'Position x')
rename_column(table_data, 'y', 'Position y')
print('fancy_header=True means that common prefixes in the keys are detected and added in a second table header:')
HTML('\n'.join(html_table(table_data, simple=True, highlight_rows=[2], fancy_header=True)))
```

Final note: if you remove `simple=True` argument from the html_table function, javacript code will be included that allows sorting of tables. This will not work inside of Jupyter though.
